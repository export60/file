```python
import json
from pyecharts import options as opts
from pyecharts.charts import Graph
```

```python
#打开weibo.json文件，将数据写入
with open('weibo.json', 'rb') as f:
    weibo = json.load(f)
    
nodes = weibo['nodes']
links = weibo['links']
categories = weibo['categories']

weibo_data = {
    'nodes': nodes, 'links': links, 'categories': categories,
    'title': '微博转发关系图',
    'layout': 'force',
    'repulsion': 50,
    'linestyle_curve': 0.2
}
```

```python
#定义创建图函数
def createGraph(data):
    graph = (
        Graph(init_opts=opts.InitOpts(height='850px'))
            .add('', data['nodes'], data['links'],
                layout=data['layout'],
                repulsion=data['repulsion'],
                categories= data['categories'],
                linestyle_opts=opts.LineStyleOpts(
                    curve=data['linestyle_curve']
                    ),
                label_opts=opts.LabelOpts(is_show=False),
                )
            .set_global_opts(
                title_opts=opts.TitleOpts(
                    title=data['title']
                ),
                legend_opts=opts.LegendOpts(is_show=False)
            )
    )
    
    return graph
```

```python
#调用函数，显示微博转发关系图
createGraph(weibo_data).render_notebook()
```
