```python
import json
from pyecharts import options as opts
from pyecharts.charts import Graph
```

```python
#数据存放在data.json格式中
with open('data.json', 'r', encoding='utf-8') as p:
    tx = json.load(p)
    nodes = tx['nodes']
    links = tx['links']
    categories = tx['categories']
    
data = {
    'nodes':nodes, 'links': links, 'categories': categories,
    'title': '人物关系图', 'layout':'circular'
}
```

```python
#定义创建图函数
def createGraph(data):
    graph = (
        Graph(init_opts=opts.InitOpts(width='1000px' ,height='850px'))
            .add('', data['nodes'], data['links'],
                layout=data['layout'],
                categories= data['categories'],
                is_rotate_label=True,
                linestyle_opts=opts.LineStyleOpts(
                    curve=0.3,
                    color='source'
                    ),
                label_opts=opts.LabelOpts(position='right'),
                )
            .set_global_opts(
                title_opts=opts.TitleOpts(
                    title=data['title']
                ),
                legend_opts=opts.LegendOpts(
                    orient='vertical',
                    pos_left='2%', pos_top='20%'
                )
            )
    )
    
    return graph
```

```python
#显示人物关系图
createGraph(data).render_notebook()
```
